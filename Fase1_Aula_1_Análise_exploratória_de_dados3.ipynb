```python
colunas_usaveis = dados.mean(numeric_only=True).index.tolist()
colunas_usaveis.insert(0, 'Unidade da Federação')
colunas_usaveis[:5]
```

#Desafio: reposicionar a legeda, retocar título da visualização e nomear eixos

```python
#Reposicionando a legenda
import matplotlib.pyplot as plt
dados_usaveis.T.plot(figsize=(10,6))
plt.legend(loc='upper right', bbox_to_anchor=(1.3,1.02))
#Incluindo título
plt.title('Dados de Saúde', fontweight='bold')
#nomeando eixos
plt.xlabel('Período',fontweight='bold', labelpad=10)
plt.ylabel('Valores',fontweight='bold', labelpad=10)
```
