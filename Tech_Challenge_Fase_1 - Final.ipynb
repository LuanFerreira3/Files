```python
#Detectando o encoding correto do arquivo
def detect_encoding(file_path):
    with open(file_path, 'rb') as f:
        raw_data = f.read()
        result = chardet.detect(raw_data)
        encoding = result['encoding']
        confidence = result['confidence']
        return encoding, confidence

file_path = '/content/ProcessaAmericanas.csv'
encoding, confidence = detect_encoding('/content/ProcessaAmericanas.csv')
print(f'O encoding do arquivo é: {encoding} com uma confiança de {confidence * 100:.2f}%')
```

```python
# Função para remover caracteres especiais, incluindo acentos
def remover_caracteres(texto):
    # Remover caracteres especiais simples
    caracteres_especiais = "',.´`^)(°+%"
    for char in caracteres_especiais:
        texto = texto.replace(char, '')
    # Remover acentos
    texto = ''.join(ch for ch in unicodedata.normalize('NFKD', texto) if not unicodedata.combining(ch))
    return texto
```

```python
# Função para remover caracteres especiais, incluindo acentos
def remover_caracteres(texto):
    # Remover caracteres especiais simples
    caracteres_especiais = "',.´`^)(°+%"
    for char in caracteres_especiais:
        texto = texto.replace(char, '')
    # Remover acentos
    texto = ''.join(ch for ch in unicodedata.normalize('NFKD', texto) if not unicodedata.combining(ch))
    return texto
```

```python
# Função para remover caracteres especiais, incluindo acentos
def remover_caracteres(texto):
    # Remover caracteres especiais simples
    caracteres_especiais = "',.´`^)(°+%"
    for char in caracteres_especiais:
        texto = texto.replace(char, '')
    # Remover acentos
    texto = ''.join(ch for ch in unicodedata.normalize('NFKD', texto) if not unicodedata.combining(ch))
    return texto
```

```python
# Função para remover caracteres especiais, incluindo acentos
def remover_caracteres(texto):
    # Remover caracteres especiais simples
    caracteres_especiais = "',.´`^)(°+%"
    for char in caracteres_especiais:
        texto = texto.replace(char, '')
    # Remover acentos
    texto = ''.join(ch for ch in unicodedata.normalize('NFKD', texto) if not unicodedata.combining(ch))
    return texto
```

```python
# Função para remover caracteres especiais, incluindo acentos
def remover_caracteres(texto):
    # Remover caracteres especiais simples
    caracteres_especiais = "',.´`^)(°+%"
    for char in caracteres_especiais:
        texto = texto.replace(char, '')
    # Remover acentos
    texto = ''.join(ch for ch in unicodedata.normalize('NFKD', texto) if not unicodedata.combining(ch))
    return texto
```

```python
# Função para remover caracteres especiais, incluindo acentos
def remover_caracteres(texto):
    # Remover caracteres especiais simples
    caracteres_especiais = "',.´`^)(°+%"
    for char in caracteres_especiais:
        texto = texto.replace(char, '')
    # Remover acentos
    texto = ''.join(ch for ch in unicodedata.normalize('NFKD', texto) if not unicodedata.combining(ch))
    return texto
```

```python
exportacao_vinho_mesa['País'] = exportacao_vinho_mesa['País'].apply(lambda x: remover_caracteres(x))
```
